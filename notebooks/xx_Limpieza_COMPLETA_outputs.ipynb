```python
# ============================================================
# BORRADO REAL DE OUTPUTS LOCALES
# ============================================================

from pathlib import Path
import shutil

PROJECT_DIR = Path("/home/jupyter/TFM_ProcesadoDeAudios")

paths_to_clean = [
    PROJECT_DIR / "data" / "eda",
    PROJECT_DIR / "data" / "diarization_input_clean_audios",
    PROJECT_DIR / "data" / "diarization_outputs",
    PROJECT_DIR / "data" / "transcription_outputs",
    PROJECT_DIR / "data" / "role_proxy_outputs",
    PROJECT_DIR / "data" / "voiceprint_outputs",
    PROJECT_DIR / "data" / "voiceprint_role_proxy_batch_outputs",
]

for p in paths_to_clean:
    if p.exists():
        shutil.rmtree(p)
        print("Borrado:", p)

# Recrear carpeta base
(PROJECT_DIR / "data").mkdir(parents=True, exist_ok=True)

print("Limpieza local terminada.")
```

```python
# ============================================================
# BORRADO DE OUTPUTS EN GCS
# ============================================================

from google.cloud import storage

bucket_name = "catedras_audio_detection"

prefixes_to_delete = [
    "pipelineA/procesados_UNAV/clean_audios/",
    "pipelineA/procesados_UNAV/diarization_outputs/",
    "pipelineA/procesados_UNAV/transcription_outputs/",
    "pipelineA/procesados_UNAV/role_proxy_outputs/",
    "pipelineA/procesados_UNAV/voiceprint_outputs/",
]

client = storage.Client()
bucket = client.bucket(bucket_name)

for prefix in prefixes_to_delete:
    blobs = list(bucket.list_blobs(prefix=prefix))
    print(f"{prefix}: {len(blobs)} archivos encontrados")

    for blob in blobs:
        blob.delete()

    print("Borrado:", prefix)

print("Limpieza GCS terminada.")
```
